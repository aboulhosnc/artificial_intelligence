```python
# In[57]:


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
get_ipython().magic(u'matplotlib inline')


# In[58]:


df = pd.read_csv('groupB.txt')
df.columns = ['Height','Weight','Sex']


# In[59]:


df


# In[60]:


df_male = df[df['Sex']==0]
df_male

df_female = df[df['Sex']==1]
df_female

sns.lmplot( x="Height", y="Weight",data=df, hue = 'Sex', palette ='GnBu', fit_reg = False)

x = np.arange(4,8)
y = (-150/7)*x+275

plt.plot(x,y,'mediumaquamarine') 
plt.show()


# In[78]:


net = (-150/7)*x+(-1)*y+275
net_output = []

for index, row in df_male.iterrows():
    x = row['Height']
    y = row['Weight']
    net = (-150/7)*x+(-1)*y+275
    #print(row['Height'], row['Weight'],net)
    net_output.append(net)
    
df_male['NEW'] = net_output
df_male.min()


# In[79]:


net = (-150/7)*x+(-1)*y+275
net_output = []

for index, row in df.iterrows():
    x = row['Height']
    y = row['Weight']
    net = (-150/7)*x+(-1)*y+275
    #print(row['Height'], row['Weight'],net)
    net_output.append(net)
    
df['NEW'] = net_output
df.min()


# In[69]:


df_male.max()


# In[70]:


df_female.max()


# In[71]:


df_threshold = 0

guess = []

for index, row in df.iterrows():
    z = row['NEW']
    if z < 0:
        guess_out = 0
        guess.append(guess_out)
    else:
        guess_out = 1
        guess.append(guess_out)

df['Guess'] = guess
df


# In[72]:


df['Check'] = df['Sex'] - df['Guess']
values = df.iloc[df['Check'].nonzero()]
print (values)


# In[75]:


false_p = 0
false_n = 0
for index, row in df.iterrows():
    z = row['Check']
    if z == -1:
        false_p = false_p + 1
    elif z == 1:
        false_n = false_n + 1
        
print(false_p, false_n)
```
