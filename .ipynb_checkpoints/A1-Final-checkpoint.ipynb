```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
#df = pd.read_csv('groupA.txt')
#df = pd.read_csv('groupB.txt')
df = pd.read_csv('groupC.txt')

df.columns = ['Height','Weight','Sex']
```

```python
df_male = df[df['Sex']==0]
df_male

df_female = df[df['Sex']==1]
df_female

sns.lmplot( x="Height", y="Weight",data=df, hue = 'Sex', palette ='GnBu', fit_reg = False)

x = np.arange(4,8)

#y = (-160/6.15)*x+285
#y = (-150/7)*x+275
y = (-140/8)*x+255

plt.plot(x,y,'mediumaquamarine') 
plt.show()
```

```python


net_output = []

for index, row in df_male.iterrows():
    x = row['Height']
    y = row['Weight']
    
    #net = (-160/6.15)*x+(-1)*y+285
    #net = (-150/7)*x+(-1)*y+275
    net = (-140/8)*x+(-1)*y+255
    
    #print(row['Height'], row['Weight'],net)
    net_output.append(net)
    
df_male['NEW'] = net_output
df_male.min()
```

```python


net_output = []

for index, row in df_female.iterrows():
    x = row['Height']
    y = row['Weight']
    
    #net = (-160/6.15)*x+(-1)*y+285
    #net = (-150/7)*x+(-1)*y+275
    net = (-140/8)*x+(-1)*y+255
    
    #print(row['Height'], row['Weight'],net)
    net_output.append(net)
    
df_female['NEW'] = net_output
df_female.min()
```

```python


net_output = []

for index, row in df.iterrows():
    x = row['Height']
    y = row['Weight']
    
    #net = (-160/6.15)*x+(-1)*y+285
    #net = ((-150/7)*x)+((-1)*y)+275
    net = (-140/8)*x+(-1)*y+255
    
    net_output.append(net)
    
df['NEW'] = net_output
df.min()
```

```python
df_threshold = df_male['NEW'].max() - df_female['NEW'].min()
df_threshold = (df_threshold / float(2)) + df_female['NEW'].min()
df_threshold
```

```python


guess = []

for index, row in df.iterrows():
    z = row['NEW']
    if z < 0:
        guess_out = 0
        guess.append(guess_out)
    else:
        guess_out = 1
        guess.append(guess_out)

df['Guess'] = guess
df
######################################
```

```python
## CHECK THE GUESSES 
df['Check'] = df['Sex'] - df['Guess']
values = df.iloc[df['Check'].nonzero()]
print values
######################################
```

```python
## COUNTING METRICS 
false_p = 0
false_n = 0
true_n = 0
true_p = 0
for index, row in df.iterrows():
    z = row['Check']
    w = row['Guess']
    if z == -1:
        false_p = false_p + 1
    elif z == 1:
        false_n = false_n + 1
    elif w == 0:
        true_n = true_n + 1
    elif w == 1:
        true_p = true_p +1
    
        
print(false_p, false_n, true_p, true_n)

#######################################
```

```python
## ACCURACY AND ERROR 
acc = 0

acc = (float(true_p + true_n)/(true_p + true_n + false_p + false_n))
print acc

err = 1-acc
print(err)
######################################
```

```python
## TRUE POSITIVE
tp = 0

tp = float(true_p)/(true_p + true_n)
print(tp)
######################################
```

```python
## TRUE NEGATIVE 
tn = 0

tn = float(true_n)/(true_p + true_n)
print(tn)
######################################
```

```python
## FALSE POSITIVE
fp = 0

fp = float(false_p)/(false_p + false_n)
print(fp)
######################################
```

```python
## FALSE NEGATIVE
fn = 0

fn = float(false_n)/(false_p + false_n)
print(fn)
######################################
```

```python
## PRECISION
prec = 0

prec = float(true_p)/(true_p + false_p)
print(prec)
######################################
```
