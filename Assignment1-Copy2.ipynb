```python
## GUESS BASED ON THRESHOLD AND NET 
df_threshold = 0

guess = []

for index, row in df.iterrows():
    z = row['NEW']
    if z < 0:
        guess_out = 0
        guess.append(guess_out)
    else:
        guess_out = 1
        guess.append(guess_out)

df['Guess'] = guess
df
######################################
```

```python
## CHECK THE GUESSES 
df['Check'] = df['Sex'] - df['Guess']
values = df.iloc[df['Check'].nonzero()]
print values
######################################
```

```python
## COUNTING METRICS 
false_p = 0
false_n = 0
true_n = 0
true_p = 0
for index, row in df.iterrows():
    z = row['Check']
    w = row['Guess']
    if z == -1:
        false_p = false_p + 1
    elif z == 1:
        false_n = false_n + 1
    elif w == 0:
        true_n = true_n + 1
    elif w == 1:
        true_p = true_p +1
    
        
print(false_p, false_n, true_p, true_n)

#######################################
```

```python
## ACCURACY AND ERROR 
acc = 0

acc = (float(true_p + true_n)/(true_p + true_n + false_p + false_n))
print acc

err = 1-acc
print(err)
######################################
```
